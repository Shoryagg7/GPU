## 1. Generate random numbers and store them in a file. Calculate CPU execution time.

This section generates a specified number of random floating-point numbers within a given range and saves them to a CSV file. It then measures the CPU time taken for this operation.

```python
import numpy as np
import pandas as pd
import time
import os

def generate_and_save_random_numbers(num_samples, min_val, max_val, filename):
    """
    Generates random numbers and saves them to a CSV file.
    Returns the CPU execution time for generation and saving.
    """
    start_time = time.process_time()

    # Generate random numbers
    random_numbers = np.random.uniform(min_val, max_val, num_samples)

    # Create a DataFrame and save to CSV
    df = pd.DataFrame({'random_number': random_numbers})
    df.to_csv(filename, index=False)

    end_time = time.process_time()
    cpu_time = end_time - start_time

    print(f"Generated {num_samples} random numbers and saved to {filename}")
    print(f"File size: {os.path.getsize(filename) / (1024*1024):.2f} MB")

    return cpu_time

# --- Configuration --- #
NUM_SAMPLES = 10_000_000  # Number of random numbers to generate
MIN_VALUE = 0.0
MAX_VALUE = 1000.0
OUTPUT_FILENAME = 'random_dataset.csv'

# --- Execute and Measure --- #
cpu_execution_time_colab = generate_and_save_random_numbers(
    NUM_SAMPLES, MIN_VALUE, MAX_VALUE, OUTPUT_FILENAME
)

print(f"CPU execution time (Google Colab): {cpu_execution_time_colab:.4f} seconds")

# --- Output for the table --- #
print("\n--- Output for Assignment Table ---")
print(f"Hardware Type: Google Colab environment")
print(f"Input file size/Matrix Size: {NUM_SAMPLES} numbers, {os.path.getsize(OUTPUT_FILENAME) / (1024*1024):.2f} MB file")
print(f"Computational Time/CPU execution Time: {cpu_execution_time_colab:.4f} seconds")
```
